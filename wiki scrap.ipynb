```python
import pandas as pd
# url = 'https://en.wikipedia.org/wiki/List_of_Singapore_MRT_stations'
url = 'https://en.wikipedia.org/wiki/List_of_Singapore_LRT_stations'

df_list = pd.read_html(url)
```

```python
df_list[1:]
```

```python
df = df_list[1]
```

```python
df.columns = df.columns.droplevel()
```

```python
df.columns = df.columns.droplevel(0)
```

```python
df.columns
```

```python
df.head()
```

```python
df['station_name'] = df['English • Malay']
# df = df.drop(['English • Malay', 'Chinese', 'Tamil',
#        'Name(s) during planning stages', 'Abbreviation', 'Location(s)',
#        'Interchange;Adjacent transportation'], axis = 1)
```

```python
df = df[['In operation' ,'station_name' ]]
```

```python
df = df[df['In operation'].str.contains("Line") == False]
```

```python
df = df.dropna(subset=['In operation'])
```

```python
df.shape
```

```python
df['station_code'] = df['In operation'].str[:4].str.strip()
```

```python
df.to_csv('lrt_data.csv', index=False)
```
